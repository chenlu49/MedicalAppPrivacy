```python
import spacy
from bs4 import BeautifulSoup

# Load the English model
nlp = spacy.load("en_core_web_sm")

# Load HTML content
path = '/Users/chenluwang/Documents/GitHub/MedicalAppPrivacy/'
with open(path + '20_www.theatlantic.com.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse HTML with BeautifulSoup
soup = BeautifulSoup(html_content, 'lxml')

# Get text from HTML
text = soup.get_text()

# Process text with spaCy
doc = nlp(text)

# Define a list of keywords that indicate conditional meaning
conditional_keywords = [
    'if', 'assuming', 'provided that', 'in case', 'unless',
    'assuming that', 'on the condition that', 'given that',
    'in the event that', 'suppose that', 'supposing that',
    'as long as', 'providing that', 'provided', 'on the assumption that',
    'if and only if', 'insofar as', 'lest'
]

# Iterate through sentences in the document
sentences = []
for sent in doc.sents:
    # Check if the sentence contains any of the conditional keywords
    if any(keyword in sent.text.lower() for keyword in conditional_keywords):  # Convert to lowercase for case-insensitive check
        print(f"Conditional Sentence: {sent.text}")
        sentences.append(sent.text)
        
        print("\n---\n")
len(sentences)
```

```python
# Define keywords related to personal information security
security_keywords = ['privacy', 'personally identifiable information', 'security', 'data protection', 'cookies', 'web beacon']

# Store sentences related to personal information security
security_related_sentences = []

# Iterate through each sentence, checking if it contains any of the keywords
for sentence in sentences:
    if any(keyword in sentence.lower() for keyword in security_keywords):  # Case-insensitive check
        security_related_sentences.append(sentence)

# Print sentences related to personal information security
for sentence in security_related_sentences:
    print(sentence)
    print("\n---\n")
len(security_related_sentences)
```

```python
# Define keywords related to the collection or use of personal information
data_keywords = ['collect', 'use', 'personal data', 'information', 'personally identifiable information', 'data', 'cookies', 'web beacon', 'store', 'sharing', 'disclosure', 'privacy policy']

# Store sentences related to the collection or use of personal information
data_related_sentences = []

# Iterate through each sentence, checking if it contains any of the keywords
for sentence in security_related_sentences:
    if any(keyword in sentence.lower() for keyword in data_keywords):  # Case-insensitive check
        data_related_sentences.append(sentence)

# Print sentences related to the collection or use of personal information
for sentence in data_related_sentences:
    print(sentence)
    print("\n---\n")
len(data_related_sentences)
```
